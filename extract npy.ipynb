```python
import torch
import torch.nn as nn
import clip
import torch.nn.functional as F
import numpy as np
from glob import glob
from transformers import CLIPTokenizerFast, CLIPProcessor, CLIPModel

# import fiftyone as fo
# import fiftyone.brain as fob
import numpy as np
from glob import glob
import json
import os
from PIL import Image
from torchvision.transforms import functional as F
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from torchvision import models
import torch
import clip
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

preprocess = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
def load_frames(frame_dir):
    frames = []
    for filename in sorted(os.listdir(frame_dir)):
        if filename.endswith(('.jpg', '.png', '.jpeg','.webp')):
            img_path = os.path.join(frame_dir, filename)
            img = preprocess(transforms.ToPILImage()(np.array(plt.imread(img_path)))).unsqueeze(0)
            frames.append(img)
    return torch.cat(frames)

def extract_features_clip(frames):
    model, preprocess_clip = clip.load("ViT-B/32", device=device)
    frames = frames.to(device)
    with torch.no_grad():
        features = model.encode_image(frames).cpu().numpy()
    return features
```

```python
keyframes_folders_all = "D:\\AIC\\videopart2\\keyframes-part2" # Keyframes_L01, Keyframes_L02,...

keyframes_videos = f'{keyframes_folders_all}\\Keyframes_L13\\keyframes' # L01_V001, L01_V002, ... 
#keyframes_per_video = f'{keyframes_folders_all}\\Keyframes_L01\\keyframes\\L01_V001'
output_dir = 'D:\\AIC\\videopart2\\clip'

for i, video in enumerate(sorted(os.listdir(keyframes_videos))):
    print(video)
    img_path = os.path.join(keyframes_videos, video)
    print(img_path)
    frame = load_frames(img_path)
    resnet_features = extract_features_clip(frame)
    # prefix_name = video[0:3]
    # postfix_index_name = str(i).zfill(3)
    np.save(os.path.join(output_dir, f'{video}.npy'), resnet_features)

print("Features extracted and saved successfully!")
```
